```python
criminal_table = df_kriminal.pivot_table(index=['year'], columns=['major_category'], values='value', aggfunc=np.sum, fill_value=0)
```

```python
criminal_table
```

```python
mpl.style.use(['ggplot'])
```

```python
criminal_last = criminal_table.tail()
criminal_last
```

## Area plot 

Membuat visualisasi area plot bedasarkan kasus kriminal 5 tahun terakhir (2012-2016)

```python
criminal_last.plot(kind='area', 
                    alpha=0.33, 
                    stacked=False, 
                    figsize=(15, 10),
                   )

plt.title('Criminal cases in the last 5 years') 
plt.ylabel('Number of cases') 
plt.xlabel('Year') 

plt.show()
```

What is the most crime cases in the last 5 years ?
* Theft and handling is the most crime cases in the last 5 years, followed by violence against the person 
* The cases of violence against the person have increased drastically since 2013


## Histogram

Membuat visualisasi histogram untuk mengetahui frequency kasus kriminal

```python
count, bin_edges = np.histogram(criminal_table, 15)
criminal_table.plot(kind ='hist', 
          figsize=(20, 7),
          bins=15,
          alpha=0.6,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen']
         )

plt.title('Histogram of Criminal cases (2008-2016)')
plt.ylabel('Years')
plt.xlabel('Number of Cases')

plt.show()
```

What is the most crime incidents from 2008 to 2016? 
* the crime with highest frequency is Theft and Handling (334054)

## Bar plot

Membuat visualisasi bar plot untuk mengetahui kasus theft and handling terbanyak pada tahun berapa

```python
criminal_theftnhandling = criminal_table['Theft and Handling']
criminal_theftnhandling.plot(kind='barh', figsize=(13, 5), color= ['gold'])
plt.xlabel('Number of cases')
plt.ylabel('Year')
plt.title('Theft and Handling cases 2008-2016')

plt.show()
```

In what year did theft and handling cases occur the most?
* Theft and handling mostly occurred in 2012 with the highest number almost 350000 cases

## Pie chart

Membuat visualisasi pie chart bedasarkan table criminal 2008-2016

```python
pie_criminal = criminal_table.transpose()
pie_criminal['total'] = criminal_table.sum()
```

```python
pie_criminal
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink','red','grey','lime']
explode_list = [0.1, 0.1, 0.1, 0.2, 0.2, 0.1, 0.1, 0.1, 0.1]

pie_criminal['total'].plot(kind='pie',
                           figsize=(15, 6),
                           autopct='%1.1f%%',
                           startangle=90,
                           shadow=True,
                           labels=None,
                           colors=colors_list,
                           pctdistance=1.12,
                           explode=explode_list 
                          )

plt.title('Crime cases (2008-2016)', y=1.12)
plt.axis('equal')
plt.legend(labels=pie_criminal.index, loc='upper left')

plt.show()
```

What were the top 3 crimes in 2016?
* The highest crime case is Theft and handling (41.3%), the 2nd highest is violence against person(24.2%), the third is burglary(11.7%)
