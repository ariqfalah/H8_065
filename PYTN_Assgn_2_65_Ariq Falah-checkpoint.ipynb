```python
y1 = df[df['YEAR BUILT'] == 1900]
y2 = df[df['YEAR BUILT'] == 2001]

print (y1)
print(y2)
```

Hypothesis Testing

Peningkatan populasi setiap tahunnya menyebabkan kebutuhan akan property di NYC meningkat, apakah unit sold pada tahun 2001(μ2) lebih banyak daripada tahun 1900(μ1)
* Null hypothesis : Unit sold pada tahun 2001 > unit sold pada tahun 1900
* Null hypothesis: μ2>μ1
* Alternative Hypthosis: μ2<μ1


```python
n1 = len(y1)
mu1 = y1["TOTAL UNITS"].mean()
sd1 = y1["TOTAL UNITS"].std()

n2 = len(y2)
mu2 = y2["TOTAL UNITS"].mean()
sd2 = y2["TOTAL UNITS"].std()
```

```python
(n1, mu1, sd1)
```

```python
(n2, mu2, sd2)
```

```python
sm.stats.ztest(y1["TOTAL UNITS"].dropna(), y2["TOTAL UNITS"].dropna(),alternative='two-sided')
```

* Karena p-value y2 yaitu tahun 2001 lebih kecil daripada y1 tahun 1900, maka kita dapat reject null hypothesis bahwa peningkatan populasi setiap tahunnya menyebabkan kebutuhan akan property di NYC meningkat.
